```python
import KMeans
import DBScan
import pandas as pd
import numpy as np
```

### Iris-dataset

```python
# read Iris dataset
df = pd.read_csv('Iris_dataset.csv')
df.describe()
# Ai : ith attribute/feature
```

```python
# convert dataset to numpy array
data = df.to_numpy()
# sample instances
df.head()
```

#### K-Means

```python
# create KMeans object
obj = KMeans.KMeans_class(data)

# find optimal K
obj.find_best_K()
```

```python
classification_accuracy = obj.run_on_K(3) # optimal K = 3
print("Classification Accuracy: ", classification_accuracy)
```

#### DBScan

```python
D = DBScan.DBScan_class(data)
epsilons = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
min_points = [3, 4, 5, 6, 7, 8, 9]
D.find_optimal_parameters(epsilons, min_points)
```

1. For iris-dataset, K=3 (found by K-Means algorithm using sse metric)
2. Here in a few epsilon-min_points pair K=3, so we select that pair which has K=3 and has maximum number of core points.
3. Number of core points increses with increase in epsilon and decrease in min_points 
4. best epsilon = 0.6, best min_points = 3
6. for this setting K = 3, and number of core points are more as compared to other

```python
classification_accuracy = D.DBScan_diven_param(0.6, 3)
print("Classification Accuracy: ", classification_accuracy)
```

### Spiral-dataset

```python
# read spiral dataset
df = pd.read_csv('spiral.csv')
df1 = df.drop(columns=['class'])
df1.describe()
# Ai : ith attribute/feature
```

```python
# convert dataset to numpy array
data = df.to_numpy()
# sample instances
df.head()
```

```python
# plot data
import matplotlib.pyplot as plt

label = data[:,2]
plt.scatter(data[:,0][label==0],data[:,1][label==0],color='b',marker='o',cmap='autumn')
plt.scatter(data[:,0][label==1],data[:,1][label==1],color='g',marker='o',cmap='autumn')
plt.xlabel('a0')
plt.ylabel('a1')
plt.title('spiral dataset')
plt.legend(['class0','class1'])
plt.show()
```

#### K-Means

```python
# create KMeans object
obj = KMeans.KMeans_class(data)

# find optimal K
obj.find_best_K()
```

```python
classification_accuracy = obj.run_on_K(6) # optimal K = 6
print("Classification Accuracy: ", classification_accuracy)
```

#### DBScan

```python
D = DBScan.DBScan_class(data)
epsilons = [0.4, 0.5, 0.6, 0.7, 0.8]
min_points = [10, 13, 16, 19, 22]
D.find_optimal_parameters(epsilons, min_points)
```

1. Number of core points increses with increase in epsilon and decrease in min_points
2. All the settings of epsilon and min_points pair create either 2 clusters or no clusters
3. Depending on the larger number of core points, we chose epsilon = 0.8, min_points=13

```python
classification_accuracy = D.DBScan_diven_param(0.8, 13)
print("Classification Accuracy: ", classification_accuracy)
```
