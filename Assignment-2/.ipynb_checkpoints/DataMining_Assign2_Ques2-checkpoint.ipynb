```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import pca
from sklearn.decomposition import PCA

# import created scripts from same folder
import EM
import Denclue
```

### Iris dataset

```python
# # read Iris dataset
df = pd.read_csv('Iris_dataset.csv')
data = df.to_numpy()
df.describe()
# Ai : ith attribute/feature
```

```python
# apply PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data[:,:-1])
label = data[:,-1]

#plot data
fig = plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0][label=='Iris-setosa'],pca_data[:,1][label=='Iris-setosa'],color='b',marker='o',cmap='YlGnBu')
plt.scatter(pca_data[:,0][label=='Iris-versicolor'],pca_data[:,1][label=='Iris-versicolor'],color='g',marker='o',cmap='YlGnBu')
plt.scatter(pca_data[:,0][label=='Iris-virginica'],pca_data[:,1][label=='Iris-virginica'],color='r',marker='o',cmap='YlGnBu')
plt.xlabel('u0')
plt.ylabel('u1')
plt.title('pca iris dataset')
plt.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```

#### EM-algorithm

```python
obj = EM.EM_class(data=pca_data, K=3, epsilon=0.003, colors=['orange', 'tab:green', 'tab:cyan'])
print_plot_for_iterations = [0,10,20,30,40,50,60,70]
mu, sigma, prob_c = obj.EM_algo(print_plot_for_iterations)
```

#### DENCLUE - Gaussian kernel pdf

```python
D_obj = Denclue.DENCLUE(data = pca_data) 
h_list = [0.1, 0.35, 0.6, 0.85, 1.1, 1.35]
D_obj.plot_pdf_in_h_range(h_list)        
```

```python
D = Denclue.DENCLUE(pca_data)
print('density attractors :' , D.algo(1.1, 0.02))
```

### Spiral dataset

#### EM algorithm

```python
# # read Iris dataset
df = pd.read_csv('spiral.csv')
data = df.to_numpy()
df.describe()
# Ai : ith attribute/feature
```

```python
label = data[:,2]
fig = plt.figure(figsize=(8,6))
plt.scatter(data[:,0][label==0],data[:,1][label==0],color='b',marker='o',cmap='autumn')
plt.scatter(data[:,0][label==1],data[:,1][label==1],color='g',marker='o',cmap='autumn')
plt.xlabel('a0')
plt.ylabel('a1')
plt.title('spiral dataset')
plt.legend(['class0','class1'])
plt.show()
data = data[:,:2]
```

```python
obj = EM.EM_class(data=data, K=5, epsilon=0.1, colors=['tab:green', 'tab:cyan', 'tab:red', 'tab:orange', 'tab:grey'])
print_plot_for_iterations = [0,10,20,30,40,50,60,70]
mu, sigma, prob_c = obj.EM_algo(print_plot_for_iterations)
```

#### DENCLUE - Gaussian kernel pdf

```python
D_obj = Denclue.DENCLUE(data = data) 
h_list = [0.3, 0.6, 0.9, 1.2, 1.5]
D_obj.plot_pdf_in_h_range(h_list)        
```

```python
D = Denclue.DENCLUE(data)
print('density attractors :' , D.algo(1.5, 0.005))
```
