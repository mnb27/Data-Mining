```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,5))
plot = fig.add_subplot(121) 

minSups = [0.005, 0.01, 0.04, 0.08]
Apriori_Times = [464.7182, 424.8332, 21.7083, 2.9837]
Eclat_Times = [113.9917, 17.7150, 17.2891, 4.8473]
FP_Times = [30.5733, 21.7429, 8.9001, 1.1585]


plot.plot(minSups, Apriori_Times, label = "Apriori")
plot.plot(minSups, Eclat_Times, label = "Eclat")
plot.plot(minSups, FP_Times, label = "FP Growth")
plot.set_xlabel('min_sup') 
plot.set_ylabel('time taken in seconds') 
plot.set_title('LIQUOR DATASET : Time taken by different algorithms vs min_sup') 
plot.legend() 
```
